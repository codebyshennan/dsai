## 1. Series: The Building Block

A Series is like a single column in a spreadsheet: a 1D array with labels (the index). Each value has a label, making it a "smart" array. DataFrames are built from Series (each column is a Series).

Use Series for: time series, categorical data, measurements, or any labeled 1D data.

````python
# Create Series from list (auto index 0, 1, 2, ...)
grades = pd.Series([90, 85, 95, 78])
print("Series from list:")
print(grades)
print(f"\ndtype: {grades.dtype}, shape: {grades.shape}")

print(grades.index)
print(grades.array)

````

**Custom index, selection, and boolean masks**

**What this code does:**
- Builds `grades_labeled` from the same four scores as before, but passes `index=['Alice', 'Bob', 'Charlie', 'David']` so the row labels are names instead of 0..3.
- Prints the whole Series, then selects one value with `grades_labeled['Charlie']` (label-based lookup, not position).
- Computes `grades_labeled >= 90`, which returns a boolean Series aligned to the index, and passes it into `[ ]` to keep only rows where the value is True—so you see Alice and Charlie.


````python
# Series with custom index (labels)
grades_labeled = pd.Series(
    [90, 85, 95, 78], 
    index=['Alice', 'Bob', 'Charlie', 'David']
    )

print("Series with custom labels:")
print(grades_labeled)

# Access by label
print(f"\nCharlie's grade: {grades_labeled['Charlie']}")

# Filter (boolean indexing)
print("\nPassing grades (>= 90):")
print(grades_labeled[grades_labeled >= 90])
````

**Series from a dictionary**

**What this code does:**
- Calls `pd.Series({...})` with student names as keys and grades as values; pandas turns keys into the index and values into the Series values.
- Prints the result—same labels as the previous cell, but the data came from a mapping instead of a list + index.


````python
# [90, 85, 95, 78]
# index=['Alice', 'Bob', 'Charlie', 'David']
grades_labeled_dict = pd.Series(    
    {
        'Alice': 90,
        'Bob': 85,
        'Charlie': 95,
        'David': 78
    }
)

print("Series with custom labels:")
print(grades_labeled_dict)
````

**Dictionary data with an explicit index order**

**What this code does:**
- Builds a population Series from a dict of city → millions, then passes `index=['New York', 'Tokyo', 'London', 'Paris']` to force row order and membership (same four cities, but ordered as listed).
- Prints the Series, then prints `population.mean()`—the arithmetic mean of the four displayed values.


````python
# Series from dictionary (keys become index)
population = pd.Series(
    {'New York': 8.4, 'London': 9.0, 'Tokyo': 37.4, 'Paris': 2.2},
    index=['New York', 'Tokyo', 'London', 'Paris']
)
print("Series from dict:")
print(population)
print(f"\nMean: {population.mean():.1f}")
````

**Build a DataFrame from column-wise lists**

## 4. Reindexing and Dropping

`reindex()` reorganizes data to match new labels; missing entries get NaN. Chain `.ffill()` or `.bfill()` to fill gaps. `drop()` removes rows or columns; use `axis=1` for columns. Assign the result or use `inplace=True`.

````python
# Reindexing
grades_s = pd.Series([85, 92, 78], index=['Alice', 'Bob', 'Charlie'])
new_idx = ['Alice', 'Bob', 'Charlie', 'David']
print("After reindex (added David):")
print(grades_s.reindex(new_idx))

sales_df = pd.DataFrame({'Revenue': [1000, 1500]}, index=['Jan', 'Mar'])
print("\nFill missing months:")
print(sales_df.reindex(['Jan', 'Feb', 'Mar']).ffill())
````

## 5. Arithmetic and Data Alignment

Pandas aligns data by index when doing arithmetic. Mismatched labels produce NaN. Use `fill_value` in `add()`, `sub()`, etc. to handle missing. `combine_first()` fills missing from a second DataFrame.

````python
# Alignment by index
s1 = pd.Series({'Mon': 100, 'Tue': 120, 'Wed': 150})
s2 = pd.Series({'Tue': 110, 'Wed': 140, 'Thu': 130})
print("Difference (aligned by index):")
print(s1 - s2)
````

**`fill_value` and `combine_first`**

**What this code does:**
- Builds two overlapping DataFrames `df1` and `df2` on different index sets.
- Prints `df1.add(df2, fill_value=0)` so missing pairs in the aligned add are treated as 0 before addition.
- Builds `primary` and `secondary` with NaNs in `primary`, then prints `primary.combine_first(secondary)` to fill NaNs from `secondary` where `primary` is missing.


````python
# fill_value and combine_first
df1 = pd.DataFrame({'A': [1, 2], 'B': [4, 5]}, index=['r1', 'r2'])
df2 = pd.DataFrame({'B': [7, 8], 'C': [10, 11]}, index=['r1', 'r3'])
print("Add with fill_value=0:")
print(df1.add(df2, fill_value=0))

primary = pd.DataFrame({'A': [1, np.nan, 3]}, index=['r1', 'r2', 'r3'])
secondary = pd.DataFrame({'A': [10, 20]}, index=['r1', 'r2'])
print("\nCombine (primary first):")
print(primary.combine_first(secondary))
````

## 6. Function Application and Mapping

`apply()` runs a function on each column (axis=0) or row (axis=1). For Series, use `map()` with a dict or function. Prefer vectorized ops (`df['A'] + df['B']`) over `apply()` when possible.

````python
# apply() on column
grades_df = pd.DataFrame({'Math': [85, 76, 92], 'Science': [92, 88, 95]})
def to_letter(score):
    return 'A' if score >= 90 else 'B' if score >= 80 else 'C'
grades_df['Math_Grade'] = grades_df['Math'].apply(to_letter)
print(grades_df)
````

**`map` on a Series**

**What this code does:**
- Creates `codes` with product codes and a `names` dict mapping each code to a product name.
- Prints `codes.map(names).tolist()`—each element is replaced by the dict value, producing `['Laptop', 'Mouse', 'Keyboard']`.


````python
# map() on Series
codes = pd.Series(['A123', 'B456', 'C789'])
names = {'A123': 'Laptop', 'B456': 'Mouse', 'C789': 'Keyboard'}
print("Mapped:", codes.map(names).tolist())
````

## 7. Sorting and Ranking

`sort_values()` orders by column; `sort_index()` by index. `rank()` assigns positions; use `method='min'`, `'first'`, `'dense'` for ties. `nlargest()`/`nsmallest()` for efficient top-N.

````python
# Create sales data for sorting examples
np.random.seed(42)
sales = pd.DataFrame({
    'Region': np.random.choice(['North', 'South', 'East'], 12),
    'Product': np.random.choice(['A', 'B', 'C'], 12),
    'Sales': np.random.randint(100, 500, 12)
})
sales.head()
````

**Sort by one or more columns**

**What this code does:**
- Prints the top 5 rows by `Sales` descending via `sort_values('Sales', ascending=False).head()`.
- Prints a multi-key sort: `sort_values(['Region', 'Sales'], ascending=[True, False])` so regions are alphabetical and within each region sales are highest-first.


````python
# Sort by values
print("Top 5 by Sales:")
print(sales.sort_values('Sales', ascending=False).head())

print("\nSort by Region, then Sales:")
print(sales.sort_values(['Region', 'Sales'], ascending=[True, False]).head())
````

**Ranking and `nlargest`**

**What this code does:**
- On `scores`, prints `rank()` and `rank(method='min')` so you see how tie-breaking changes the rank numbers.
- Prints `sales.nlargest(3, 'Sales')`—the three rows with the largest `Sales` without sorting the full table.


````python
# Ranking
scores = pd.Series([85, 92, 85, 95])
print("Rank (default):", scores.rank().tolist())
print("Rank (min):", scores.rank(method='min').tolist())
print("\nTop 3 with nlargest:")
print(sales.nlargest(3, 'Sales'))
````
