### Setup (imports and style)

**Purpose:** Load `pyplot` and `numpy`, and apply a consistent Seaborn-like style for all figures below.

**Walkthrough:** `plt.style.use('seaborn')` aliases to the Matplotlib seaborn-v0_8 style when available; run this cell once before the plots.


```python
# Import the libraries we need
import matplotlib.pyplot as plt
import numpy as np

# Make our charts look nice
plt.style.use('seaborn')
```

## 1. Line Chart: Daily Temperature

**Purpose:** Plot weekday labels against temperatures so you can read day-to-day change at a glance.

**Walkthrough:** `plot` with `marker='o'`, `xticks(rotation=45)`, `grid`, and `tight_layout` cover the usual readability fixes.

Let's create a simple line chart showing temperature changes over a week.

```python
# Our data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
temperatures = [75, 72, 78, 71, 76]

# Create the line chart
plt.figure(figsize=(10, 6))  # Make the chart a bit bigger
plt.plot(days, temperatures, marker='o')  # Add markers at each point

# Add labels and title
plt.title('Daily Temperature', fontsize=14)
plt.xlabel('Day of Week')
plt.ylabel('Temperature (F)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()  # Adjust spacing
plt.show()
```

## 2. Bar Chart: Fruit Sales

**Purpose:** Compare discrete categories (fruit) with labeled bar heights and per-bar color.

**Walkthrough:** `bar` returns artists; the loop sets colors and `text` annotates each bar top.

Now let's make a bar chart comparing fruit sales.

```python
# Our data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sales = [45, 62, 38, 50]

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(fruits, sales)

# Add colors
colors = ['#ff9999', '#ffcc99', '#ff9933', '#9933ff']
for bar, color in zip(bars, colors):
    bar.set_color(color)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

# Customize the chart
plt.title('Fruit Sales', fontsize=14)
plt.xlabel('Fruit Type')
plt.ylabel('Number Sold')

plt.show()
```

## 3. Pie Chart: Daily Activities

**Purpose:** Show shares of a 24-hour day that sum to 100%—only when "part of whole" is the question.

**Walkthrough:** `autopct` labels wedges; `legend` with `bbox_to_anchor` places keys beside the pie.

Let's see how someone spends their day using a pie chart.

```python
# Our data
activities = ['Sleep', 'Work', 'Free Time', 'Other']
hours = [8, 8, 5, 3]

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(hours, labels=activities, autopct='%1.1f%%',
        colors=['lightblue', 'lightgreen', 'pink', 'orange'])

# Add title
plt.title('How I Spend My Day', fontsize=14)

# Add legend
plt.legend(activities, title="Activities", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

## 4. Scatter Plot: Height vs Weight

**Purpose:** Explore correlation between two numeric variables with jittered synthetic data.

**Walkthrough:** `np.random.seed` fixes the sample; `scatter` uses `alpha` to handle overlap; axes are labeled in physical units.

Let's create a scatter plot to show the relationship between height and weight.

```python
# Generate some sample data
np.random.seed(42)  # For reproducibility
heights = np.random.normal(170, 10, 50)  # 50 heights around 170cm
weights = heights * 0.4 + np.random.normal(50, 5, 50)  # Weights loosely based on heights

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(heights, weights, alpha=0.6, c='purple')

# Customize the chart
plt.title('Height vs Weight', fontsize=14)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

## 5. Multiple Plots: Dashboard Style

**Purpose:** Reuse variables from earlier cells in a 2×2 grid with a single figure title.

**Walkthrough:** `plt.subplots` unpacks four axes; prior cells must be run so `days`, `temperatures`, `fruits`, `sales`, `colors`, `activities`, `hours`, `heights`, and `weights` exist; `suptitle` is figure-level.

Let's combine multiple plots into a single figure.

```python
# Create a figure with subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Line plot
ax1.plot(days, temperatures, marker='o', color='blue')
ax1.set_title('Temperature Over Week')
ax1.set_xticklabels(days, rotation=45)

# 2. Bar plot
ax2.bar(fruits, sales, color=colors)
ax2.set_title('Fruit Sales')

# 3. Pie chart
ax3.pie(hours, labels=activities, autopct='%1.1f%%')
ax3.set_title('Daily Activities')

# 4. Scatter plot
ax4.scatter(heights, weights, alpha=0.6, c='purple')
ax4.set_title('Height vs Weight')
ax4.grid(True)

# Adjust layout
plt.tight_layout()

# Add a main title
fig.suptitle('Data Visualization Examples', fontsize=16, y=1.02)

plt.show()
```
